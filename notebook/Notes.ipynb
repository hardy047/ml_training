#### Rules for prediction:

Factors are which that we thought can influence our prediction<br>
1)	Factors should be quantifiable and measurable <br>
2)	Historical evidence for factors (multiple evidence)<br>
3)	Time lag between factors and outcomes should be in consistence with problem statement<br>
4)	History should repeat itself. (If context is change or factors itself changing then historical data can’t be use)<br>

Number of evidence = number of factors * 10

If outcome is numerical it called regression problem.<br>
If outcome is categorical it called classification problem.<br>

Supervised Learning -> Regressoin and Classification<br>
Unsupervised Learning -> Clustering<br>

Y = f(x1, x2, ... xn)<br>
x is Factors/features/Independent variables<br>
y is outcome,target or dependent variables<br>

Variance = 1/N Sum(x1 - X{Mean})2<br>
correlation = SUM((x1 - X{Mean})*(Y1 - Y{MEAN})/std(x)*std(y))<br>
std = Standard Deviation<br>

For all regression model try to reduce MSE. (Mean squared error)

### Process of building Regression model
1. Identify outcome variable an feature variable
2. If any categorical features then create dummy variable
3. Split data in training and test data
4. Build a model using training date using cross validation
5. Check accuraccy using R squared
6. If R squared is not good do following
    a. Feature engineering
    b. use different model 
    c. add more factors
7. Repeat 4 to 6 steps.
8. Select a model and test it against test dataset
9. Calculate RMSE

### Process of building Classification model
1. Identify outcome variable an feature variable
2. If any categorical features then create dummy variable
3. Split data in training and test data
4. Build a model check confusion matrix
5. Identify which qundrants are important to business and assign cost
6. Calculate cut off probability using cost
7. Test it against test data
8. Calculate presion, recall and f1 factor
